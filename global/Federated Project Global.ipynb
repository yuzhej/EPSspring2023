```python
import pandas as pd
```

```python
data = pd.read_csv('COVID-19_Case_Surveillance_Public_Use_Data_with_Geography.tsv', sep ='\t')
```

```python
data.shape
```

```python
col_to_drop=['res_county','county_fips_code','case_positive_specimen_interval','case_onset_interval','underlying_conditions_yn']
```

```python
df = data.drop(col_to_drop, axis='columns')
```

```python
df.shape
```

```python
100*(df.isnull().sum())/len(df)
```

```python
df.dtypes
```

```python
df['value'] = 1
```

```python
df.dtypes
```

```python
import numpy as np
```

```python
table = pd.pivot_table(df, values='value', index=['state_fips_code', 'res_state'],
                       columns=['case_month'], aggfunc=np.sum, fill_value=0)
```
